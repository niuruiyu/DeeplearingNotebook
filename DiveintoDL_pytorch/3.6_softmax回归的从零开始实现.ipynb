```python
import torch
import torchvision
import numpy as np
import sys
import d2lzh_pytorch as d2l
import torchvision.transforms as transforms
```

```python
#获取和读取数据
#这里和书中有差别，书中使用了自己封装的新的函数
#这里仍然使用dataset和dataloader
batch_size = 256
mnist_train = torchvision.datasets.FashionMNIST(root = '~/DataSets/FashionMNIST',train = True,download = True,transform = transforms.ToTensor())
mnist_test = torchvision.datasets.FashionMNIST(root = '~/DataSets/FashionMNIST',train = False,download = True,transform = transforms.ToTensor())
import sys
if sys.platform.startswith('win'):
    num_workers = 0
else:
    num_workers = 4
train_iter = torch.utils.data.DataLoader(mnist_train,batch_size,shuffle=True,num_workers=num_workers)
test_iter = torch.utils.data.DataLoader(mnist_test,batch_size,shuffle=True,num_workers= num_workers)
```

上面导入数据集完成
## 2.初始化模型参数
输入长度为28*28=784，图像有10个类别，所以softmax的权重为784*10，bias为1*10

```python
num_inputs = 784
num_outputs = 10
w = torch.tensor(np.random.normal(0,0.01,(num_inputs,num_outputs)),dtype = torch.float)
b = torch.zeros(num_outputs,dtype = torch.float)
```

```python
#需要模型参数梯度
w.requires_grad_(True)
b.requires_grad_(True)
print(w.requires_grad)
```

## 3.实现softmax运算
下面先对tensor的按维度操作进行介绍

```python
newx = torch.tensor([[1,2,3],[4,5,6]])
#对同一列（dim=0）的数字进行求和，保留行和列（keepdim=True）这两个维度
print(newx.sum(dim=0,keepdim=True))
#对同一行（dim=1）的数字进行求和，保留维度
print(newx.sum(dim=1,keepdim=True))
```

定义sofhmax函数

```python
def softmax(X):
    X_exp = X.exp()
    partition = X_exp.sum(dim=1,keepdim=True)
    return X_exp/partition
```

## 4.定义模型

```python
def net(X):
    return softmax(torch.mm(X.view((-1,num_inputs)),w)+b)
```

## 5.定义损失函数
使用前面说的交叉熵损失函数，可以使用gather函数得到样本的预测概率

```python
def cross_entropy(y_hat,y):
    return -torch.log(y_hat.gather(1,y.view(-1,1)))
```

## 5.计算分类准确率
得到预测的概率分布之后，会把里面概率最大的类别作为输出类别  
如果这个类别和真是类别一样，那就是预测正确  
分类准确率就是“预测正确的数量”/“总预测数量”

```python
def accuracy(y_hat,y):
    return (y_hat.argmax(dim=1) ==y).float().mean().item()
```

y_hat.argmax(dim=1)获取y——hat每行中最大元素的索引  
==之后就可以获得一个类型为ByteTensor的Tensor，用float（）转换为1，0数值，计算平均值（其实也就是准确率），用item获得准确率的数值

可以把上面拓展为评价模型net在数据集data_iter上的准确率函数
在d2lzh——pytorch文件中添加了evaluate_accuracy函数

```python
print(d2l.evaluate_accuracy(test_iter,net))
```

前面我们初始化了参数，所以实际准确率会很低

## 6.训练模型
使用的是小批量随机梯度下降来优化模型的损失函数

```python
num_epochs ,lr = 5,0.1
def train_ch3(net,train_iter,test_iter,loss,num_epochs,batch_size,params=None,lr=None,optimizer=None):
    for epoch in range(num_epochs):
        #一个epoch的送loss，acc，以及样本数的初始化
        train_l_sum,train_acc_sum,n = 0.0,0.0,0
        for X,y in train_iter:
            y_hat = net(X)
            l = loss(y_hat,y).sum()
            #梯度清零,在反向传播之前清掉之前计算中累计的梯度
            if optimizer is not None:#当优化器传入了对象时，自动清零所有通过优化器管理的参数
                optimizer.zero_grad()
            elif params is not None and params[0].grad is not None:#当没有传入优化器，但是传入了参数列表时，先检查参数列表不为空且第一个参数有梯度，然后遍历所有的参数，对所有的参数的梯度进行清零
                for param in params:#手动清零
                    param.grad.data.zero_()
            #反向传播
            l.backward()
            if optimizer is None:
                d2l.sgd(params,lr,batch_size)#默认sgd方法
            else:
                optimizer.step()
            train_l_sum+=l.item()
            train_acc_sum+=(y_hat.argmax(dim=1) ==y).sum().item()
            n += y.shape[0]
        test_acc = d2l.evaluate_accuracy(test_iter,net)
        print('epoch %d,loss%.4f,train acc %.3f,test acc %.3f'
              %(epoch+1,train_l_sum/n,train_acc_sum/n,test_acc))
train_ch3(net,train_iter,test_iter,cross_entropy,num_epochs,batch_size,[w,b],lr)
            
        
```
