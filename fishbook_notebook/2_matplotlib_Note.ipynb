```python
#绘制sin函数
import numpy as np
import matplotlib.pyplot as plt
#生成数据
x = np.arange(0,6,0.1)
y = np.sin(x)
#绘制图形
plt.plot(x,y)
plt.show()
```

```python
#追加cos函数
x = np.arange(0,6,0.1)
y = np.sin(x)
y2 = np.cos(x)
#绘制图形
plt.plot(x,y,label='sin')
plt.plot(x,y2,linestyle='--',label='cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin&cos')
plt.legend()
plt.show()
```
